```python
# Import required libraries
import numpy as np
import tensorflow as tf

# Step 1: Create a small random dataset
np.random.seed(42)  # For reproducible results
input_data = np.random.rand(4, 5) * 10  # 4 samples, 5 features #(rows and column)

print("Original Input Values:\n")
print(input_data)

# Step 2: Apply Batch Normalization
batch_norm = tf.keras.layers.BatchNormalization() #keras → Deep Learning API ,,,,layers → Contains neural network layers

# training=True ensures batch statistics are used
batch_output = batch_norm(input_data, training=True)

print("\nBatch Normalization Output:\n")
print(batch_output.numpy())

# Step 3: Apply Layer Normalization
layer_norm = tf.keras.layers.LayerNormalization()

layer_output = layer_norm(input_data)

print("\nLayer Normalization Output:\n")
print(layer_output.numpy())

# Step 4: Short Comparison
print("\nComparison:")
print("--------------------------------------------")
print("1. Batch Normalization normalizes the input using the mean and variance of the entire batch.")
print("2. It helps stabilize and speed up the training of deep neural networks.")
print("3. Layer Normalization normalizes the features within each individual sample.")
print("4. It works independently of the batch size and is commonly used in RNNs and Transformers.")
print("5. Difference: Batch Normalization depends on batch statistics, whereas Layer Normalization depends only on the features of each sample.")
print("--------------------------------------------")
```
