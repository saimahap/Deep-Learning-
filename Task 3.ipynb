```python
import ollama

# Get input from the user
prompt = input("Enter your prompt: ")

# List of unsafe keywords
unsafe_words = [
    "ignore",
    "system prompt",
    "jailbreak",
    "bypass",
    "developer mode"
]

# Assume the prompt is safe
safe = True

# Check if any unsafe keyword is present
for word in unsafe_words:
    if word in prompt.lower():
        safe = False
        print("\n⚠ Potential Prompt Injection Detected!")
        print("Reason: Unsafe instruction found.")
        print("Request not sent to the LLM.")
        break

# If safe, send to Ollama
if safe:
    response = ollama.chat(
        model="llama3.2",
        messages=[
            {"role": "user", "content": prompt}
        ]
    )

    print("\nResponse:")
    print(response["message"]["content"])
```
